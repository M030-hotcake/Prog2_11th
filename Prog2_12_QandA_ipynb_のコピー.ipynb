# 問題1（初級）

1～100までの自然数の中から0％～100％の確率(自然数)でランダムに数字が当たるプログラミングを考える。

この試行を100回繰り返し、試行回数、当選確率、あたりかはずれの3つをcsvファイルに書き込む。

その後、横軸を試行回数、縦軸を当選確率(%)と当選回数(その試行回数までの総和)としてグラフを描画せよ。(濵野)

* 問題提案者の氏名を問題末尾に記入して下さい．複数人で作成した場合は連名で構いません．

```python
# ここに解答または解答例を記入する
import csv
import random
import matplotlib.pyplot as plt
with open('sample.csv','w',encoding='UTF-8',newline='')as f:
  date=[]
  for n in range(100):
    ran=random.randint(0,100)
    count=0
    win=[]
    while True:
      if count==ran:
        break
      rand=random.randint(1,100)
      if rand in win:
        continue
      else:
        win.append(rand)
        count=count+1
    if n+1 in win:
      judge='あたり'
    else:
      judge='はずれ'
    date2=n+1,ran,judge
    date.append(date2)
  write=csv.writer(f)
  write.writerows(date)
with open('sample.csv','r',encoding='UTF-8',newline='')as g:
  read=csv.reader(g)
  date_x=[]
  date_y1=[]
  date_y2=[]
  count2=0
  for i in read:
    date_x.append(int(i[0]))
    date_y1.append(int(i[1]))
    if i[2]=='あたり':
      count2=count2+1
    date_y2.append(count2)
  plt.plot(date_x,date_y1)
  plt.plot(date_x,date_y2)
  plt.show()
```

## 問題1解説
random関数を用いてあたりとなる数字の数を決め、1～100までの自然数の中からrandom関数を用いて、被りなくあたりとなる数字を決め、あたりとなる数字(解答例はその時の試行回数の数をあたりとしているがこれもrandom関数を道いてもよい)が選んだ自然数の中にあるかを判定し、試行回数、当選確率、あたりかはずれの3つをcsvファイルに書き込む。

その後は、csvファイルに書き込んだ3つのデータを基にグラフを描画することで完了する。(濵野)

# 問題3（上級）

5人の生徒の1年～3年の数学の模試の点数を格納したscores.csvがある。下記のプログラムを読み込んで、それぞれの点数の推移をひとつのグラフにして出力しなさい。(林)．

```python
%%file scores2.csv
name,1st,2nd,3rd
Alice,53,67,79
Bob,57,72,73
Charlie,89,66,71
Dave,55,31,57
Ellen,34,47,49
```

```python
# ここに解答または解答例を記入する
import csv
import matplotlib.pyplot as plt
with open('scores2.csv','r',encoding='utf-8',newline='') as f:
  data=[x for x in csv.reader(f)]
alice=[int(data[1][i]) for i in range(1,4)]
bob=[int(data[2][i]) for i in range(1,4)]
chalie=[int(data[3][i]) for i in range(1,4)]
dave=[int(data[4][i]) for i in range(1,4)]
ellen=[int(data[5][i]) for i in range(1,4)]
x=1,2,3
plt.xticks([ 1,2,3 ])
plt.plot(x,alice,label='Alice')
plt.plot(x,bob,label='Bob')
plt.plot(x,chalie,label='Chalie')
plt.plot(x,dave,label='Dave')
plt.plot(x,ellen,label='Ellen')
plt.legend()
plt.show()
```
